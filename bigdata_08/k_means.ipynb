```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array([[6,3], [11,15], [17,12], [24,10], [20,25], [85,70], [71,81], [60,79], [56,52], [81,91], [80,81]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.show()
```

```python
n_cluster = range(1, 10)
kmeans = [KMeans(n_clusters=i) for i in n_cluster]
score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))]
```

```python
plt.plot(n_cluster, score)
plt.xlabel('Number of Cluster')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
iris_data
```

```python
X = iris_data.data
Y = iris_data.target
iris = pd.concat([pd.DataFrame(X), pd.DataFrame(Y)], axis=1)
iris.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
iris.head()
```

```python
fig, ax = plt.subplots()
for name, group in iris.groupby('Species'):
    ax.scatter(group.SepalLength, group.SepalWidth, marker='o', label=name)
    ax.legend(fontsize=10, loc='lower right')
```

```python
fig, ax = plt.subplots()
for name, group in iris.groupby('Species'):
    ax.scatter(group.PetalLength, group.PetalWidth, marker='o', label=name)
    ax.legend(fontsize=10, loc='lower right')
```
